```python
#create a function to get price action of any market, exchange and period combination 
#can potentially add period to the input variables for dynaic
def get_price_action_1d(exchange, market):
    candles=cw.markets.get("{}:{}".format(exchange, market)
                           , ohlc=True
                           , periods=["1d"])
    df = pd.DataFrame(candles.of_1d)
    df.columns=(["date_utc","open", "high", "low", "close", "volume", "volume in {}".format(market[-3:])])
    df.set_index("date_utc", inplace=True)
    return df

print("Function defined sucessfully")
```

```python
df_btc_usd_kraken = get_price_action_1d(exchange="KRAKEN"
                                    ,market = "BTCUSD"
                                    )

df_btc_usd_kraken.head()
```
